* 방의 개수, 거리만 이상치 정리
  * 제거하지 않고 평균값으로 대치
* Postcode와 Propertycount는 RobustScaler

```python
train_num.describe()
```

```python
train_num.describe().loc["75%", "Rooms"]
```

```python
room_iqr = train_num.describe().loc["75%", "Rooms"] - train_num.describe().loc["25%", "Rooms"]
dist_iqr = train_num.describe().loc["75%", "Distance"] - train_num.describe().loc["25%", "Distance"]

room_iqr, dist_iqr
```

```python
condition = train_num["Rooms"] > train_num.describe().loc["75%", "Rooms"] + room_iqr * 1.5
train_num.loc[condition, "Rooms"] = train_num.describe().loc["mean", "Rooms"]
```

```python
condition = train_num["Distance"] > train_num.describe().loc["75%", "Distance"] + room_iqr * 1.5
train_num.loc[condition, "Distance"] = train_num.describe().loc["mean", "Distance"]
```

```python
from sklearn.preprocessing import RobustScaler

rs_post = RobustScaler()
train_num[["post_rs"]] = rs_post.fit_transform(train_num[["Postcode"]])
train_num[["Postcode", "post_rs"]].head()
```

```python
rs_property = RobustScaler()
train_num[["property_rs"]] = rs_property.fit_transform(train_num[["Propertycount"]])
train_num[["Propertycount", "property_rs"]].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

```python
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
models = {
    'SVR':SVR(),
    'XGBRegressor':XGBRegressor(),
    'Ridge':Ridge(),
    'ElasticNet':ElasticNet(),
    'SGDRegressor':SGDRegressor(),
    'BayesianRidge':BayesianRidge(),
    'LinearRegression':LinearRegression(),
    'RandomForestRegressor':RandomForestRegressor()
}
```

```python
model_results = []
model_names = []

# model 하나씩 순회
for name, model in models.items():
    # 모델 학습
    a = model.fit(x_train, y_train)
    # 모델 예측
    predicted = a.predict(x_val)
    # 모델 rmse로 평가
    score = np.sqrt(mean_squared_error(y_val, predicted))
    # 평가 저장
    model_results.append(score)
    model_names.append(name)
    
# 저장된 평가 데이터프레임 만들기
df_results = pd.DataFrame([model_names, model_results])
df_results = df_results.transpose()
# RMSE 기준 내림차순 정렬
df_results = df_results.rename(columns={0:'Model', 1:'RMSE'}).sort_values(by='RMSE', ascending=False)

print(df_results)
```
